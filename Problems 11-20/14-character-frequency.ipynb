### সমস্যার বিবরণ

এমন একটি প্রোগ্রাম লিখতে হবে, যেটি একটি বাক্যে কোনো একটি বিশেষ বর্ণ কত বার আছে, সেটি প্রিন্ট করবে।

### ইনপুট

প্রথম লাইনে একটি পূর্ণসংখ্যা ${T (1 <= T <= 100)}$ থাকবে। পরবর্তীতে $(2 \ x \ T)$-সংখ্যক লাইন থাকবে। পরবর্তী প্রতি এক জোড়া লাইনের প্রথমটিতে একটি স্ট্রিং (string) থাকবে, যার দৈর্ঘ্য সর্বোচ্চ ${10,000}$ এবং দ্বিতীয় লাইনে একটি ক্যারেক্টার (character) থাকবে।

### আউটপুট

প্রোগ্রামটির আউটপুটে ক্যারেক্টারটি স্ট্রিংয়ে কত বার আছে, সেটি প্রিন্ট করতে হবে নিম্নরূপে।

### নমুনা ইনপুট / আউটপুট

![](https://i.ibb.co/2ZCXJWG/image.png)

````python
T = int(input())

for _ in range(1, T+1):
    string = input()
    character = input()
    
    count = 0
    for i in string:
        if i == character:
            count += 1
        
    if count == 0:
        print(f"'↬ {character}' is not present")
    else:
        print(f"↬ Occurrence of '{character}' in '{string}' = {count}")
````

### C++ Solution

```cpp
#include <bits/stdc++.h>
using namespace std;

int main() {
   int T;
   cin >> T;
   cin.ignore();
   
   for (int i = 1; i <= T; i++) {
      string str;
      getline(cin, str);

      char character;
      cin >> character;
      cin.ignore();
      
      int count = 0;
      for (char ch: str) {
         if (tolower(character) == tolower(ch)) {
            count += 1;
         }
      }
      
      if (count == 0) {
          cout << "'" << character << "'" << " is not present" << endl;
      } else {
         cout << "Occurrence of " << "'" << character << "'" << " in " << "'" << str << "'" << " = " << count << endl;
      }
   }
}
```
